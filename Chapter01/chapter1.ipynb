```python
# 用这一列的中位数填充
df_bikes['windspeed'].fillna((df_bikes['windspeed'].median()), inplace=True)
```

```python
df_bikes.iloc[[56, 81],:]
```

```python
# 使用groupby()进行分组，并使用每组的中位数进行填充
df_bikes.select_dtypes(include=['int64','float64']).groupby(['season']).median()
```

```python
# 使用中位数填充
# transform()可以应用于整个 DataFrame 的一列
df_bikes['hum'] = df_bikes['hum'].fillna(df_bikes.groupby('season')['hum'].transform('median'))
```

```python
# 获取特定特征值位na的行
df_bikes[df_bikes['temp'].isna()]
```

```python
# 用前后两天的平均值来填充缺失值
mean_temp = (df_bikes.iloc[700]['temp'] + df_bikes.iloc[702]['temp'])/2
mean_atemp = (df_bikes.iloc[700]['atemp'] + df_bikes.iloc[702]['atemp'])/2
df_bikes['temp'].fillna((mean_temp), inplace=True)
df_bikes['atemp'].fillna((mean_atemp), inplace=True)
```

```python
df_bikes.dropna(how='any', inplace=True)
```

```python
df_bikes[df_bikes.isna().any(axis=1)]
```

```python
# 将时间列转为日期格式
df_bikes['dteday'] = pd.to_datetime(df_bikes['dteday'])
```

```python
df_bikes['dteday'] = df_bikes['dteday'].apply(pd.to_datetime, infer_datetime_format=True, errors='coerce')
```

```python
import datetime as dt
```

```python
# 提取月份信息
df_bikes['mnth'] = df_bikes['dteday'].dt.month
```

```python
df_bikes.tail()
```

```python
df_bikes.loc[730, 'yr'] = 1.0
```

```python
df_bikes.tail()
```

```python
df_bikes = df_bikes.drop('dteday', axis=1)
```

```python
df_bikes = df_bikes.drop(['casual', 'registered'], axis=1)
```

```python
# 划分特征与标签
X = df_bikes.iloc[:,:-1]
y = df_bikes.iloc[:,-1]
```

```python
# 划分训练集和测试集
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df_bikes[df_bikes.isna().any(axis=1)]
```

```python
# 线性回归实例化
lin_reg = LinearRegression()
# 进行训练
lin_reg.fit(X_train, y_train)
# 进行预测
y_pred = lin_reg.predict(X_test)

from sklearn.metrics import mean_squared_error
import numpy as np
# 计算误差
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("RMSE: %0.2f" % (rmse))
```

下面用XGBoost进行回归预测


```python
from xgboost import XGBRegressor
#首先进行实例化, 再进行训练，再进行预测。计算均方误差和均方根误差

xg_reg = XGBRegressor()
xg_reg.fit(X_train, y_train)
y_pred = xg_reg.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("RMSE: %0.2f" % (rmse))
```

```python
# 使用带交叉验证的线性回归
from sklearn.model_selection import cross_val_score
model = LinearRegression()
scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=10)
rmse = np.sqrt(-scores)

#十折交叉验证返回的是一个列表
print('Reg rmse:', np.round(rmse, 2))
print('RMSE mean: %0.2f' % (rmse.mean()))
```

```python
# 设置优化目标位最小化均方误差
model = XGBRegressor(objective="reg:squarederror")
scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=10)

rmse = np.sqrt(-scores)
print('Reg rmse:', np.round(rmse, 2))
print('RMSE mean: %0.2f' % (rmse.mean()))
```

```python
# 传入人口普查收入数据
df_census = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data')
# Display first 5 rows
df_census.head()
```

```python
# 发现表格的第一行并不是表头, 而是直接就是数据，所以设置header=None
df_census = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header=None)
df_census.head()
```

```python
# 设置表头
df_census.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation',
                  'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 
                   'income']
df_census.head()
```

```python
df_census.info()
```

```python
df_census = df_census.drop(['education'], axis=1)
```

```python
# 进行one-hot编码
df_census = pd.get_dummies(df_census)
df_census.info()
```

```python
for i in df_census.select_dtypes(include='bool').columns:
    df_census[i]=(df_census[i]).astype('int')
```

```python
df_census.info()
```

```python
# Drop column 'income_ <=50K'
df_census = df_census.drop('income_ <=50K', axis=1)
```

```python
# 划分特征与标签
X = df_census.iloc[:,:-1]
y = df_census.iloc[:,-1]
X
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
# 定义的交叉验证函数，参数为分类器和折数
def cross_val(classifier, num_splits=10):
    model = classifier
    scores = cross_val_score(model, X, y, cv=num_splits)
    print('Accuracy:', np.round(scores, 2))
    print('Accuracy mean: %0.2f' % (scores.mean()))
```

```python
# 采用逻辑回归的分类法
cross_val(LogisticRegression())
```

```python
from xgboost import XGBClassifier
cross_val(XGBClassifier())
```
